# Data Studying

## 01 Fd or Firearms dataset

```python
df_fd.info()
#determine shape of data
```

```python
print(df_fd['Deaths'].value_counts())
#display values of deaths columns
```

```python
def check_if_suppressed(value):
    try:
        float(value)
        return 'Numeric'
    except ValueError:
        return 'Suppressed'
#create function to determine if item is numeric or suppressed
```

```python
df_fd['Death_Category'] = df_fd['Deaths'].apply(check_if_suppressed)
#apply function to dataframe
```

```python
print(df_fd['Death_Category'].value_counts())
#see new column
```

```python
df_fd['Year'] = df_fd['Year'].dt.year
#convert year column to integer
```

```python
df_fd.isnull().sum()
#determine any missing values
```

```python
df_fd['Deaths'] = pd.to_numeric(df_fd['Deaths'], errors='coerce')
#change suppressed values to NaN 
```

```python
df_pivot = df_fd.pivot_table(
    index=['State', 'Year', 'Gender', 'ICD-10 113 Cause List'], 
    columns='Five-Year Age Groups', 
    values='Deaths', 
    aggfunc='sum'
).reset_index()
#pivot dataset to display age groups as columns instead of rows
```

```python
pd.set_option('display.max_columns', None)
#increase size of displayable columns
```

```python
df_pivot.head(5)
#display data change
```

```python
df_pivot.describe()
#study statistical analysis of firearms dataset
```
